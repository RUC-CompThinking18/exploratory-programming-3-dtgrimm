line 7 searches through the string being passed through beowulf and finds all words that end with the characters "at"
line 11 adds words that are greater than three characters long to the list results

```python
import re
source = open('pg16328.txt')
beo = source.read()
def beowulf(aString):
    if type(aString) != str: 
        raise TypeError("Not a string") 
    result = re.findall(r'\w*at\b', aString)
    x = []
    for element in result:
        if len(element) > 3:
            x.append(element)
    return x
            
```

```python
beowulf(beo)
```
